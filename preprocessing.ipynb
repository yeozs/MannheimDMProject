```python
data = pd.read_csv('airlines_reviews.csv')
df=pd.DataFrame(data)
df.head(5)
```

```python
# group dataset according to their ariline, and collect all reviews
df_grouped=df.groupby("Airline")
df_grouped["Reviews"].apply(lambda x: ' '.join(x)).reset_index(name="Reviews")
```

```python
def get_valid_tokens(tokens: Union[List[str], StreamBackedCorpusView], remove_stopwords: bool=False) -> List[str]:
    """
    :param tokens: list of tokens that should be cleaned
    :param remove_stopwords: bool indicating if stopwords should be removed 
                             False by default
    :return: list of valid tokens
    """
    # your code here
    valid_tokens = []
    stop_words = set(nltk.corpus.stopwords.words('english'))
    for token in tokens:
        token = token.lower()
        if token not in string.punctuation and not token.isdigit():
            # Optionally, remove stopwords
            if not remove_stopwords or token not in stop_words:
                valid_tokens.append(token)
    return valid_tokens
```

```python
# store in txt
for i in list (df_grouped.groups.keys()):
    with open("{}.txt".format(i), 'w', encoding='utf-8') as f:
        for value in df_grouped.get_group(i)["Reviews"]:
             f.write(f"{value}\n")
```
